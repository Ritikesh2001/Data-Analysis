## Data Analysis

In this step, we answer key analytical questions such as:
- Average final grade
- Number of students scoring above 15
- Relationship between study time and grades
- Gender-wise performance comparison

```python
# Average Grade

df['G3'].mean()
```

```python
# How many students scored above 15?  

len(df[df['G3'] > 15])
```

```python
df['studytime'].corr(df['G3'])
```

```python
df.groupby('sex')['G3'].mean()
```

## Data Visualization

Visualizations are created to better understand data distribution
and relationships between variables.

```python
# Histograms for grade

plt.hist(df['G3'])
plt.show()
```

```python
plt.scatter(df['studytime'], df['G3'])
plt.show()
```

```python
df.groupby('sex')['G3'].mean().plot(kind = 'bar')
plt.show()
```

## Conclusion

- The average final grade of students is approximately **10.42**.
- **40 students** scored above 15 marks in the final examination.
- There is a slight positive relationship between study time and academic performance.
- **Male students** performed slightly better on average compared to female students.
